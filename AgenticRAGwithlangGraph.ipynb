```python
import gradio as gr
from langchain import hub
from langchain.output_parsers import PydanticOutputParser
from langchain.document_loaders import WebBaseLoader
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings,
)
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_core.prompts import PromptTemplate
from langchain_community.document_loaders import PyMuPDFLoader,SitemapLoader
from langchain.vectorstores import FAISS
from langchain.document_loaders import ReadTheDocsLoader
import transformers
import requests
from langchain.document_loaders import UnstructuredExcelLoader,UnstructuredURLLoader
from langchain_community.document_loaders import WebBaseLoader
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from transformers import AutoTokenizer, pipeline
from langchain import PromptTemplate, HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain.chains import LLMChain
import os
from langchain.chains import RetrievalQA
from langgraph.graph import StateGraph, END
from langchain_core.output_parsers import JsonOutputParser
from typing import Dict, TypedDict, Optional
from pprint import pprint
from langchain.schema import Document
from typing import List
from langchain_core.prompts import ChatPromptTemplate
faiss_db = 'vectorstore/db_faiss'
loader = PyMuPDFLoader("/content/Netflix Goes to Bollywood.IC__1.pdf")
data = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(data)
embedding_function = SentenceTransformerEmbeddings(model_name="all-mpnet-base-v2")
db1 = FAISS.from_documents(data, embedding_function)


#def process_excel_file(excel_path):
    # Load the Excel file using UnstructuredExcelLoader
loader = UnstructuredExcelLoader('/content/Netflix-Movies-Data-for-Analysis.xlsx', mode="elements")
document = loader.load()


    # Initialize a text splitter to divide documents into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
texts = text_splitter.split_documents(document)



embedding_function = SentenceTransformerEmbeddings(model_name="all-mpnet-base-v2")

    # Create a vector store using FAISS from the text chunks and embeddings
db2 = FAISS.from_documents(texts, embedding_function)


web_links = ["https://medium.com/@sauravgosain1/how-has-netflix-changed-the-cinema-industry-cc139879eb43",]
loader = WebBaseLoader(web_links)


docs = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits1 = text_splitter.split_documents(docs)
embedding_function = SentenceTransformerEmbeddings(model_name="all-mpnet-base-v2")
db3 = FAISS.from_documents(docs, embedding_function)
db1.merge_from(db2)
db1.merge_from(db3)

retriever=db1.as_retriever( search_type="similarity_score_threshold",search_kwargs={'score_threshold': 0.8})
text_generation_pipeline = transformers.pipeline(
   model=model,
   tokenizer=tokenizer,
   task="text-generation",
   temperature=0.2,
   repetition_penalty=1.1,
   return_full_text=True,
   max_new_tokens=300,
)


llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
prompt_template= """
 Answer the question based only on the following context:
{context}
You are allowed to rephrase the answer based on the context.
Question: {question}
"""
prompt = ChatPromptTemplate.from_template(prompt_template)
# Post-processing
# def format_docs(docs):
#     return "\n\n".join(doc.page_content for doc in docs)

# Chain
rag_chain = prompt | llm | StrOutputParser()

# Prompt

class GraphState(TypedDict):
    question : str
    generation : str
    length: int
    classification:str
    documents : List[str]
def classify_input_node(state):
    question = state["question"]
    if(question=="Hi"):
       return {"classification":"greeting"}
    else:
       return {"classification":"question"}
def handle_greeting_node(state):
    return {"generation": "Hello! How can I help you today?"}


def handle_RAG(state):
    """
    Generate answer using RAG on retrieved documents

    Args:
        state (dict): The current graph state

    Returns:
        state (dict): New key added to state, generation, that contains LLM generation
    """
    # print("---GENERATE--")
    question = state["question"]
    documents = retriever.invoke(question)
    # RAG generation
    generation = rag_chain.invoke({"context": documents, "question": question})

    return { "question": question, "generation": generation}

workflow = StateGraph(GraphState)
workflow.add_node("classify_input", classify_input_node)
workflow.add_node("handle_greeting", handle_greeting_node)
workflow.add_node("handle_RAG", handle_RAG)

workflow.set_entry_point("classify_input")
workflow.add_edge('handle_greeting', END)
workflow.add_edge("handle_RAG", END)
def decide_next_node(state):
    return "handle_greeting" if state.get('classification') == "greeting" else "handle_RAG"
workflow.add_conditional_edges(
    "classify_input",
    decide_next_node,
    {
        "handle_greeting": "handle_greeting",
        "handle_RAG": "handle_RAG"
    }
)


app = workflow.compile()
from pprint import pprint
def get_important_facts(question):
 inputs = {"question": question}
 for output in app.stream(inputs):
  for key, value in output.items():
        pprint(f"Finished running: {key}:")
 return(value["generation"])
iface = gr.Interface(
  fn=get_important_facts,
  inputs=gr.Textbox(lines=2, placeholder="Enter your question here..."),
  outputs="text",
  title="RAG with MoE",
  description="Ask questions about the provided context",
)

# Launch the Gradio app
iface.launch(debug=True)
```
